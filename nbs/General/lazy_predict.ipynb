```python
import lazypredict
from lazypredict.Supervised import LazyClassifier
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

## Data Manipulation

```python
data = load_breast_cancer()
data
```

```python
data.data.shape
```

```python
data.target.shape
```

```python
data.feature_names
```

```python
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

## Lazy Classifier Model

```python
clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
```

```python
print(models)
```

```python
models
```

## Lazy Regressor

```python
from lazypredict.Supervised import  LazyRegressor
from sklearn.utils import shuffle
from sklearn import datasets
import numpy as np
```

```python
boston = datasets.load_boston()
boston.keys()
```

```python
X, y = shuffle(boston.data, boston.target, random_state=42)
X.shape, y.shape
```

```python
offset = int(X.shape[0] * 0.9)
```

```python
X_train, X_test = X[offset:], X[:offset]
len(X_train), len(X_test)
```

```python
y_train, y_test = y[offset:], y[:offset]
len(y_train), len(y_test)
```

```python
reg  = LazyRegressor(custom_metric=None, ignore_warnings=False, verbose=0)
models, predictions  = reg.fit(X_train, X_test, y_train, y_test)
```

```python
models
```
