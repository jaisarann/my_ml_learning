## Analyse Year related feature

```python
year_features = [feature for feature in numerical_feature if 'yr' in feature.lower() or 'year' in feature.lower()]
year_features
```

```python
df1 = df.copy()
for feature in year_features:
    if feature != 'YrSold':
        df1[feature] = df1['YrSold'] - df1[feature]
        plt.scatter(df1[feature], df1['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
        
```

### Numerical variables has 2 types 1) Contineous 2) Discrete

## Analyse Discrete variables

```python
Disc_Num_features = [feature for feature in numerical_feature if len(df[feature].unique()) < 25 and feature not in year_features]
len(Disc_Num_features), Disc_Num_features
```
